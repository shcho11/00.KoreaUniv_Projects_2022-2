```python
file = open('drive/MyDrive/20221111_chosunilbo_opinion.txt')
reply_text = file.readlines()
file.close()
print(reply_text)
```

```python
# 1) 불용어 제거를 하지 않은 버전 

okt = Okt()
text = ''
for sentence in reply_text : 
  for noun in okt.nouns(sentence) : 
    text += noun + ' '

print(text)
```

### 1)에서와 같이 불용어 제거를 하지 않은 결과, '데', '이', '그', '것', '전', '배' 등과 같은 목적상 불필요한 단어 요소들이 결과에 포함이 됨. 아래에서 이것들을 제거해 보자.

```python
okt = Okt()
text1 = ''
for sentence in reply_text : 
  for noun in okt.nouns(sentence) : 
    if noun in ['데','이','그','것','건','전','사','배','등','중','못','수','미분'] : 
      pass
    else : 
     text1 += noun + ' '

print(text1)
```

### 2) 에서와 같이 불용어 제거를 하고 난 후 한결 결과가 정제됨.
- 아래 이미지는 tableau에서 수행한 결과물

20221111_wordcloud2.png

```python
from PIL import Image 
mask_image = np.array(Image.open('drive/MyDrive/cloud_image5.jpg'))

wc = WordCloud(background_color = 'black',
               font_path = 'drive/MyDrive/HMFMMUEX.TTC',
               max_words = 100,
               max_font_size = 80,
               min_font_size = 10,
               mask = mask_image,
               colormap = 'terrain_r').generate(text1)
wc.to_file('wc.png')
```
